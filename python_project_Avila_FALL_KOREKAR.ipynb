### This function creates a new data frame containing only values from a single writer

```python
def sub_df(classe, df):
    return [df.iloc[_] for _ in range(len(df)) if df.iloc[_]['class'] == classe ]
```

### With this function, we create a new dataframe containing the average of the 3 features choosen per writer.

```python
def moy_frame(df):
    classes = ['A','B','C','D','E','F','G','H','I','W','X','Y']
    columns_name = ['moy_row', 'moy_intercol', 'moy_interline']
    subA = pd.DataFrame({}, index = classes, columns = columns_name)
    
    for _ in classes :
        sub = pd.DataFrame(sub_df(_, df))
        moy_row = sub['row_num'].sum() / len(sub)
        moy_ic = sub['intercolumn'].sum() / len(sub)
        moy_il = sub['interline'].sum() / len(sub)
        
        data = {
            'moy_row': [[moy_row]],
            'moy_intercol': [[moy_ic]],
            'moy_interline':[[moy_il]]
        }
       # tmp = pd.DataFrame([[moy_row, moy_ic, moy_il]], columns = columns_name)
        tmp = pd.DataFrame(data)
        subA.append(tmp)
        
    return subA
```

# Machine Learning

```python
from sklearn.ensemble import GradientBoostingClassifier, RandomForestClassifier, AdaBoostClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix, roc_curve, auc, recall_score, precision_score, accuracy_score
from sklearn.preprocessing import Normalizer
from sklearn.decomposition import PCA
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis

from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import confusion_matrix
```

```python
X_train = avila_training_set[["intercolumn","upp_margin","low_margin","exploitation","row_num","mod_ratio","interline","weight","peak_num","mod_ratio/interline"
]]
y_train = avila_training_set["class"]
X_test = avila_testing_set[["intercolumn","upp_margin","low_margin","exploitation","row_num","mod_ratio","interline","weight","peak_num","mod_ratio/interline"
]]
y_test = avila_testing_set["class"]
```

```python
normalizer = Normalizer()
X = pd.DataFrame(normalizer.fit_transform(X_train))
```

### We create a list to store results of accuracies

```python
acc =[]
```

```python
#Classifier
model = KNeighborsClassifier(n_neighbors=1)
model.fit(X_train, y_train)
y_model = model.predict(X_test)

from sklearn.metrics import accuracy_score
acc.append(accuracy_score(y_test, y_model))
```

```python
mat = confusion_matrix(y_test, y_model)

seaborn.heatmap(mat, square=True, annot=True, cbar=False)
plt.xlabel('predicted value')
plt.ylabel('true value');
```

```python
#LDA
lda = LinearDiscriminantAnalysis(n_components=3)
X_trainl = lda.fit(X_train, y_train).transform(X_train)
X_testl = lda.transform(X_test)
lda.fit(X_trainl,y_train)
l_pred = lda.predict(X_testl)

#Accuracy
acc.append(accuracy_score(y_test, l_pred))
```

```python
mat = confusion_matrix(y_test, l_pred)

seaborn.heatmap(mat, square=True, annot=True, cbar=False)
plt.xlabel('predicted value')
plt.ylabel('true value');
```

```python
# Random forest
classifierRF = RandomForestClassifier()
classifierRF.fit(X_train,y_train)
pred = classifierRF.predict(X_test)

#Accuracy
acc.append(accuracy_score(y_test,pred))
```

```python
mat = confusion_matrix(y_test, pred)

seaborn.heatmap(mat, square=True, annot=True, cbar=False)
plt.xlabel('predicted value')
plt.ylabel('true value');
```

```python
#Gaussian
model = GaussianNB()                       
model.fit(X_train, y_train)        
y_model = model.predict(X_test)

#Accuracy
acc.append(accuracy_score(y_test, y_model))
```

```python
mat = confusion_matrix(y_test, y_model)

seaborn.heatmap(mat, square=True, annot=True, cbar=False)
plt.xlabel('predicted value')
plt.ylabel('true value');
```
